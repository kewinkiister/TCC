```python
import numpy as np
import matplotlib
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt
import pandas as pd
import math 
pd.set_option("display.max_rows", None, "display.max_columns", None)


df = pd.read_csv ('Dados_Kewin/ExportData1.csv')
a=df.values;
Dir = df.iloc[:, 3:4];
Dir=Dir.values
Dir = Dir.reshape(-1,15);

y = ["3","5","7","9","11", "13", "15", "10", "11", "12", "13", "14"]
x = ["0.6","0.8", "1.0", "1.2","1.4", "1.6", "1.8", "2.0", "1.4", "1.5", "1.6", "1.7", "1.8", "1.9", "2.0"]



```

```python
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cax = divider.append_axes("right", size="5%", pad=0.05)
    cbar = ax.figure.colorbar(im,ax=ax, **cbar_kw )
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom",fontsize= 18,fontname= 'Times New Roman')

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="major", color="w", linestyle='-', linewidth=0)
    ax.tick_params(which="minor", bottom=False, left=False)
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)

    return im, cbar


```

```python
fig, ax = plt.subplots(figsize=(5,5))

im, cbar = heatmap(Dir, y, x, ax=ax,
                   cmap="jet", cbarlabel="Directive(dB)")
ax.set_title("Lunenburg Toroidal Lens", fontname= 'Times New Roman', fontsize=18)
plt.ylabel('D1/Lambda', fontname= 'Times New Roman', fontsize=18)
plt.xlabel('D1/D2', fontname= 'Times New Roman', fontsize=18)
fig.tight_layout()
plt.show()
fig.savefig('plot.pdf')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = ["","","","6","","","","10","","", "","14"]
y = ["","","0.8","", "","", "1.2","","","", "1.6","", "", "","2.0"]

def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="",xlabel="",ylabel="", **kwargs):
    if not ax:
        ax = plt.gca()
        
    im = ax.imshow(data, **kwargs)
    
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    #cbar(location='left')
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom",
                       fontname= 'Times New Roman', fontsize=18 )
    
    
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)
    
    ax.tick_params(top=False, bottom=False,
                   labeltop=True, labelbottom=False)
    plt.setp(ax.get_xticklabels(), ha="right",
             rotation_mode="anchor")
    
    plt.setp(ax.get_xticklabels(),fontname= 'Times New Roman', fontsize=18)
    plt.setp(ax.get_yticklabels(),fontname= 'Times New Roman', fontsize=18)
        
    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="major", color="w", linestyle='-', linewidth=0)
    ax.tick_params(which="minor", bottom=False, left=False)
    plt.xlabel(xlabel, labelpad =-265)
    plt.setp(ax.set_ylabel(ylabel),fontname= 'Times New Roman', fontsize=18)
    plt.setp(ax.set_xlabel(xlabel),fontname= 'Times New Roman',fontsize=18)
    
    return im, cbar    
  
fig, ax = plt.subplots()   
im, cbar = heatmap(Dir, x, y, ax=ax,
                   cmap="jet", cbarlabel="Directive(dB)",
                   xlabel="D1/D2", ylabel="D1/ λ")

fig.tight_layout()
plt.show()
fig.savefig('plot.png', dpi=1000)
```

```python
import pandas as pd
base = pd.read_csv('original.csv')
base.head()
```

```python
x=base.iloc[:, 5:6].values
y=base.iloc[:, 2:3 ].values
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
x = scaler_x.fit_transform(x)
scaler_y = StandardScaler()
y = scaler_y.fit_transform(y)
print(base.columns)
```

```python
import tensorflow as tf
colunas_usadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
print(colunas_usadas)
base=pd.read_csv('original.csv', usecols = colunas_usadas)
base.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
x = base.drop('price', axis =1 )
y = base.price
```

```python
x.head()
```

```python
y.head()
```

```python
previsores_colunas = colunas_usadas[1:17]
previsores_colunas
```

```python
import tensorflow as tf
```

```python
colunas = [tf.feature_column.numeric_column(key = c) for c in previsores_colunas]
```

```python
print(colunas[0])
```

```python
from sklearn.model_selection import train_test_split
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x,y,test_size=0.3)
```

```python
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(
    x = x_treinamento,y = y_treinamento,batch_size=32, 
    num_epochs=None, shuffle=True)
```

```python
teste = tf.estimator.inputs.pandas_input_fn(
    x = x_teste,y = y_teste,batch_size=32, 
    num_epochs=10000, shuffle=False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
regressor.train(input_fn=funcao_treinamento,steps=10000)
```

```python
metricas_treinamento=regressor.evaluate(input_fn=funcao_treinamento, steps=10000)
```

```python
metricas_teste=regressor.evaluate(input_fn=teste, steps=10000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

```python
funcao_previsao=tf.estimator.inputs.pandas_input_fn(x=x_teste
        ,shuffle = False)
```

```python
previsores=regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsores)
```

```python
valores_previsoes=[]
for p in regressor.predict(input_fn=funcao_previsao):
    valores_previsoes.append(p['predictions'])
```

```python
import numpy as np
valores_previsoes = np.asarray(valores_previsoes).reshape(-1,1)
```

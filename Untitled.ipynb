```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

df = pd.read_csv ('Dados_Kewin/90.csv')
a=df.values;
Theta = df.iloc[:, 4:5];
Theta = Theta.values;
Gain = df.iloc[:, 5:6];
Gain = Gain.values;
#Gain = 10*np.log(Gain);
N_amostras=1801;
Theta = Theta.reshape(-1, N_amostras);
Gain = Gain.reshape(-1, N_amostras);
```

```python
def find_3db_opening(array1, array2):
    value= array1.max()-3
    array1 = np.asarray(array1)
    idx = (np.abs(array1 - value)).argmin()
    return 2*(90-array2[idx])
```

```python
def find_SLL(array1):
    i=1;
    j=1;
    SSL=[]
    SSLM=0
    while i<=(len(array1)-2):
        if(array1[i-1]<array1[i] and array1[i]>array1[i+1]):
            SSL.append(array1[i])                    
        i=i+1;       
    max1=max(SSL);
    SSL.remove(max1);
    max2=max(SSL);
    SSLM=max1-max2;
        
    return SSLM
```
